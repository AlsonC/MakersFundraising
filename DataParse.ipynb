```python
# prompt: for each page, add all of the pages onto a single dataframe. then for each row concatonate all the responses into one joined field called 'joined' which is a new column at the end of the dataframe, do this with every page in the staff_data.xlsx excel. Finally, download this dataframe

from google.colab import drive
import pandas as pd
from google.colab import files

drive.mount('/content/drive')

# Assuming 'staff_data.xlsx' is in your MyDrive
file_path = '/content/drive/MyDrive/staff_data.xlsx'

try:
    # Read the Excel file into a pandas DataFrame
    xls = pd.ExcelFile(file_path)
    all_data = pd.DataFrame()

    for sheet_name in xls.sheet_names:
      df = pd.read_excel(xls, sheet_name=sheet_name)
      all_data = pd.concat([all_data, df], ignore_index=True)

    # Concatenate responses into a single 'joined' column
    all_data['joined'] = all_data.astype(str).apply(' '.join, axis=1)

    # Save the combined DataFrame to a new Excel file
    all_data.to_excel('combined_staff_data.xlsx', index=False)

    # Download the combined Excel file
    files.download('combined_staff_data.xlsx')

except FileNotFoundError:
    print(f"Error: File not found at {file_path}")
except Exception as e:
    print(f"An error occurred: {e}")
```
